```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = [1,2,3]
Y = [1,2,3]
```

```python
plt.plot(X, Y,marker = "o")
```

Our loss function, which we want to minimize is

$J(\beta) = \frac{1}{2N} \sum^{N}_{i=1}{(Y_i - \beta X_i)^2}$

```python
def loss_func(b, X, Y):
    N = len(X)
    ss = np.sum( (np.array(Y) - b * np.array(X))**2 )
    result = 1/(2*N) * ss
    return result
```

```python
loss_func(b=2, X=X, Y=Y)
```

Now we want to compute the total cost across three different values of beta

```python
betas = [0, 0.5, 1, 2]
costs = [loss_func(b=b, X=X,Y=Y) for b in betas]
print(betas)
print(costs)
plt.plot(betas, costs, marker = "o")
```
