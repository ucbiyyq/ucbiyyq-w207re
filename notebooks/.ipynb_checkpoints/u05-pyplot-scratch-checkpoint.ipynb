```python
import numpy as np
import matplotlib.pyplot as plt
```

https://matplotlib.org/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py

plot a single plot

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

plot a grid of plots

```python
def my_plotter(ax, data):
    out = ax.plot(data)
    return out

data1, data2, data3, data4 = np.random.randn(4, 100)
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1)
my_plotter(ax2, data2)
```

plot list of plots automatically?

yes but only if the number of rows in the data matrix matches the number of columns in the figure

```python
def my_plotter(ax, data):
    out = ax.plot(data)
    return out

dat_mat = np.random.randn(4, 100)
fig, axs = plt.subplots(1, 4)

for ax, dat in zip(axs, dat_mat):
    my_plotter(ax, dat)
```

Can I do that without for loop?

yes but it has this awkward list coercion thing that I haven't figured out

```python
def my_plotter(ax, data):
    out = ax.plot(data)
    return out

dat_mat = np.random.randn(4, 100)
fig, axs = plt.subplots(1, 4)

list(map(my_plotter, axs, dat_mat))
```
