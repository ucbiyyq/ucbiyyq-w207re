# 1 
Create a 10x10 grid to visualize 10 examples of each digit. Python hints:

- plt.rc() for setting the colormap, for example to black and white
- plt.subplot() for creating subplots
- plt.imshow() for rendering a matrix
- np.array.reshape() for reshaping a 1D feature vector into a 2D matrix (for rendering)

```python
def P1(num_examples=10):

    # helper function to display a single digit, 
    # given tuple of, an axis from pyplot and
    # an array representing the greyscale pixel values
    def plt_dgt(ax, pxs, sh = (img_px, img_px)):
        pxs_r = np.reshape(pxs, sh)
        out = ax.imshow(pxs_r, cmap='Greys')
        return out
    
    # helper function to display multiple examples in a row,
    # given an array of axes from pyplot,
    # and given a 1 by n matrix, where each row is an array of pixels,
    def plt_exp(axr, pxss):
        for ax, pxs in zip(axr, pxss):
            plt_dgt(ax, pxs)
    
    # define the number of examples we want for each digit
    num_examples = 10
    
    # for each digit 0 to 9, finds first 10 examples
    examples = dict()
    for x in range(0, num_examples):
        fltr = np.where(mini_train_labels.astype(int) == x)[0][0:num_examples]
        examples[x] = mini_train_data[fltr]
    
    # plot our 10 by 10 matrix
    fig = plt.figure()
    fig, axs = plt.subplots(10, num_examples, figsize=(12,12))
    for r, pxss in zip(axs, examples.values()):
        plt_exp(r, pxss)
    
P1(10)
```

# 2

Evaluate a K-Nearest-Neighbors model with k = [1,3,5,7,9] using the mini training set. Report accuracy on the dev set. For k=1, show precision, recall, and F1 for each label. Which is the most difficult digit?

- KNeighborsClassifier() for fitting and predicting
- classification_report() for producing precision, recall, F1 results

```python
def P2(k_values):
    classifiers = map(lambda k: KNeighborsClassifier(n_neighbors=k), k_values)
    fits = map(lambda clsfr: clsfr.fit(mini_train_data, mini_train_labels), classifiers)
    fits = list(fits)
    accuracies = map(lambda clsfr: clsfr.score(dev_data, dev_labels), fits)
    accuracies = list(accuracies)
    
    # prints the k values and accuracies
    print(k_values)
    print(accuracies)
    
    # prints the report on the k = 1 fitted classifier
    preds = fits[0].predict(dev_data)
    print(classification_report(preds, dev_labels))
    
    # for visuals, plots the accuracies
    plt.plot(k_values, accuracies, marker = "o")
    plt.title("k = 1 has the best overall basic accuracy")

k_values = [1, 3, 5, 7, 9]
P2(k_values)
```

ANSWER: Based on the f1 scores for each digit for the k = 1 fitted classifier, it looks like the digit 8 is the most difficult to predict

# 3

Using k=1, report dev set accuracy for the training set sizes below. Also, measure the amount of time needed for prediction with each training size.

- time.time() gives a wall clock value you can use for timing operations

```python
def P3(train_sizes, accuracies):
    
    # helper function to calculate accuracy,
    # and time duration for fitting the classifier
    def time_score(clsfr, td, tl):
        start_time = time.time()
        clsfr.fit(td, tl)
        duration = time.time() - start_time
        accuracy = clsfr.score(dev_data, dev_labels)
        return (accuracy, duration)
    
    # helper function to map all classifiers against the given data and label
    def map_classifers(classifiers, td, tl):
        a_d = map(lambda clsfr: time_score(clsfr, td, tl), classifiers)
        return a_d
    
    # calculates accuracies and times for all train sizes, given k = 1
    k_values = [1]
    classifiers = list(map(lambda k: KNeighborsClassifier(n_neighbors=k), k_values))
    train_data_subsets = map(lambda x: train_data[:x], train_sizes)
    train_label_subsets = map(lambda x: train_labels[:x], train_sizes)
    results = list(map(lambda td, tl: map_classifers(classifiers, td, tl), train_data_subsets, train_label_subsets))
    
    # unpacks results
    a_d_list = []
    [a_d_list.append(*a_d) for a_d in results]
    accuracies, times = zip(*a_d_list)
    
    print("train_sizes")
    print(train_sizes)
    print("accuracies")
    print(accuracies)
    print("times")
    print(times)
    
    plt.plot(train_sizes, accuracies, marker = "o")
    plt.title("basic accuracy increases as training size increases")
    plt.show()
    
    plt.plot(train_sizes, times, marker = "o")
    plt.title("time increases as training size increases, usually")
    plt.show()

train_sizes = [100, 200, 400, 800, 1600, 3200, 6400, 12800, 25000]
# train_sizes = [100, 200, 400, 800]
accuracies = []
P3(train_sizes, accuracies)
```

# 5

Fit a 1-NN and output a confusion matrix for the dev data. Use the confusion matrix to identify the most confused pair of digits, and display a few example mistakes.

- confusion_matrix() produces a confusion matrix

```python
def P5():
    # helper function to display a single digit
    def plt_dgt(pxs, sh = (img_px, img_px)):
        pxs_r = pxs.reshape(sh)
        plt.figure(figsize=(1,1))
        plt.imshow(pxs_r, cmap='Greys')
        plt.show()
    
    # fits a 1-NN classifier to the mini-train set
    clsfr = KNeighborsClassifier(n_neighbors=1)
    clsfr.fit(mini_train_data, mini_train_labels)
    preds = clsfr.predict(dev_data)
    
    # calculates and prints the confusion matrix
    cm = confusion_matrix(dev_labels, preds)
    print("confusion matrix for 1-NN")
    print(cm)
    print("looks like 4-9 is the most confused pair")
    
    # from the cm, identify the indexes of the most confused pair
    are_fours = dev_labels.astype(int) == 4
    pred_nines = preds.astype(int) == 9
    idx_confused_pairs = np.where(are_fours & pred_nines)[0]
    
    # prints out 3 of the mismatches
    for i in idx_confused_pairs[0:3]:
        print("true label: {}".format(dev_labels[i]))
        print("prediction: {}".format(preds[i]))
        plt_dgt(dev_data[i])

P5()
```

A common image processing technique is to smooth an image by blurring. The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian -- that is, the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

Implement a simplified Gaussian blur by just using the 8 neighboring pixels: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values. Try applying your blur filter in 3 ways:
- preprocess the training data but not the dev data
- preprocess the dev data but not the training data
- preprocess both training and dev data

Note that there are Guassian blur filters available, for example in scipy.ndimage.filters. You're welcome to experiment with those, but you are likely to get the best results with the simplified version I described above.

```python
def P6():
    from scipy.ndimage.filters import gaussian_filter
    
    # helper function to display a single digit
    def plt_dgt(pxs, sh = (img_px, img_px)):
        pxs_r = pxs.reshape(sh)
        plt.figure(figsize=(1,1))
        plt.imshow(pxs_r, cmap='Greys')
        plt.show()
    
#     # try out the gaussian filter from scipy
#     t1 = mini_train_data[1]
#     t2 = gaussian_filter(input=t1, sigma=1)
#     plt_dgt(t1)
#     plt_dgt(t2)
    
#     # diy blur?
#     def myblur(pxs):
#         test = pxs.reshape((28,28))
#         test2 = test.reshape((1,784))
#         return(test2)
        
#     t1 = mini_train_data[1]
#     t2 = myblur(t1)
#     plt_dgt(t1)
#     plt_dgt(t2)
    
    # preps the various blurred data
    blurred_train_data = list(map(lambda vec: gaussian_filter(input=vec, sigma=1), mini_train_data))
    blurred_dev_data = list(map(lambda vec: gaussian_filter(input=vec, sigma=1), dev_data))
    
    # creates a 1-NN classifier
    clsfr = KNeighborsClassifier(n_neighbors=1)
    
    # blurs training data, but not the dev data
    clsfr.fit(blurred_train_data, mini_train_labels)
    preds = clsfr.predict(dev_data)
    print("blurred training data but not the dev data")
    print(classification_report(preds, dev_labels))
    
    # blurs dev data, but not the training data
    clsfr.fit(mini_train_data, mini_train_labels)
    preds = clsfr.predict(blurred_dev_data)
    print("blurred dev data but not the training data")
    print(classification_report(preds, dev_labels))
    
    # blurs both training and dev data
    clsfr.fit(blurred_train_data, mini_train_labels)
    preds = clsfr.predict(blurred_dev_data)
    print("blurred both training and dev data")
    print(classification_report(preds, dev_labels))
    
P6()
```

ANSWER: Comparing F1 scores, 1-NN predictions most accurate if both train and dev data is blurred. Applying the blur to the dev data only seems to perform the worst out of the three methods.

Fit a Naive Bayes classifier and report accuracy on the dev data. Remember that Naive Bayes estimates P(feature|label). While sklearn can handle real-valued features, let's start by mapping the pixel values to either 0 or 1. You can do this as a preprocessing step, or with the binarize argument. With binary-valued features, you can use BernoulliNB. Next try mapping the pixel values to 0, 1, or 2, representing white, grey, or black. This mapping requires MultinomialNB. Does the multi-class version improve the results? Why or why not?

```python
def P7():
    # helper function to display a single digit
    def plt_dgt(pxs, sh = (img_px, img_px)):
        pxs_r = pxs.reshape(sh)
        plt.figure(figsize=(3,3))
        plt.imshow(pxs_r, cmap='Greys')
        plt.show()
        
    # helper function for making bins
    def b_w_g(pxs, th = [0.1, 0.9]):
        pxs_b_w_g = np.copy(pxs)
        pxs_b_w_g[pxs_b_w_g < th[0]] = 0 # white
        pxs_b_w_g[pxs_b_w_g >= th[1]] = 2 # black
        pxs_b_w_g[(pxs_b_w_g >= th[0]) & (pxs_b_w_g < th[1])] = 1 # grey
        return pxs_b_w_g
    
#     # test the binning function
#     t1 = dev_data[0]
#     plt_dgt(t1)
#     plt.hist(t1, bins = 25)
#     plt.show()
#     t2 = b_w_g(t1)
#     plt_dgt(t2)
#     plt.hist(t2, bins = 25)
#     plt.show()
    
    
    # binary
    # binarize=0.0 treats any feature with a value greater than 0 as 1, otherwise 0
    clsfr = BernoulliNB(binarize=0.0)
    clsfr.fit(mini_train_data, mini_train_labels)
    preds = clsfr.predict(dev_data)
    accuracy = clsfr.score(dev_data, dev_labels)
    
    print("accuracy and f1 scores for BernoulliNB with binarize=0.0")
    print(accuracy)
    print(classification_report(preds, dev_labels))
    
    # multinomial data
    # bins the feactures of the train data
    th = [0.5, 0.9]
    binned_train_data = list(map(lambda pxs: b_w_g(pxs, th), mini_train_data))
    clsfr = MultinomialNB()
    clsfr.fit(binned_train_data, mini_train_labels)
    preds = clsfr.predict(dev_data)
    accuracy = clsfr.score(dev_data, dev_labels)
    
    print("accuracy and f1 scores for MultinomialNB with pixels binned at", th)
    print(accuracy)
    print(classification_report(preds, dev_labels))

P7()
```

ANSWER: MultinomialNB with binning, depending on the thresholds chosen, generally performs worse. For thresholds [0.5, 0.9], it looks like the classifier has better f1 scores for simpler digits like 0 and 1, but suffer on more complex digits like 8 and 9. Possibly the BernoulliNB classifer performs better because it is like a form of smoothing, whereas putting the pixels into three bins is introducing more noise.

# 8

Use GridSearchCV to perform a search over values of alpha (the Laplace smoothing parameter) in a Bernoulli NB model. What is the best value for alpha? What is the accuracy when alpha=0? Is this what you'd expect?

- Note that GridSearchCV partitions the training data so the results will be a bit different than if you used the dev data for evaluation.

```python
def P8(alphas):
    
    # creates the bernouli classifier
    clsfr = BernoulliNB(binarize = 0.0)
    
    # passes the classifier into the grid search object, then tries to fit the alphas
    gscv = GridSearchCV(estimator = clsfr, param_grid = alphas, cv = 5, iid = False)
    gscv.fit(mini_train_data, mini_train_labels)
    
    # plots the grid scores to visuzally show the best one
    mean_scores = gscv.cv_results_["mean_test_score"]
    print(alphas["alpha"])
    print(mean_scores)
    plt.figure(figsize=(8,4))
    plt.plot(alphas["alpha"], mean_scores, marker = "o")
    plt.title("alpha 0.001 seems to have the best accuracy")
    plt.xlabel('alphas')
    plt.ylabel('mean_test_scores')
    plt.show()
    
    # returns the GridSearchCV object for use later
    return gscv

alphas = {'alpha': [0.0, 0.0001, 0.001, 0.01, 0.1, 0.5, 1.0, 2.0, 10.0]}
# alphas = {'alpha': [0.0, 0.0001, 0.001, 0.01, 0.1, 0.5, 1.0, 2.0]}
nb = P8(alphas)
```

```python
print(nb.best_params_)
```

ANSWER: BernoulliNB with binarize = 0.0 and alpha 0.001 has the best accuracy, at 82.5%. An alpha of 0 gives us only an accuracy of 81.3%. After alpha 0.001, the more smoothing we add, the worse the accuracy. I suppose past a certain point, the more smoothing we do, the more information we are discarding.

- The estimated probability of each pixel is stored in feature\_log\_prob\_. You'll need to use np.exp() to convert a log probability back to a probability.

```python
def P10(num_examples):
    
    # helper function to display a single digit, 
    # given tuple of, an axis from pyplot and
    # an array representing the greyscale pixel values
    def plt_dgt(ax, pxs, sh = (img_px, img_px)):
        pxs_r = np.reshape(pxs, sh)
        out = ax.imshow(pxs_r, cmap='Greys')
        return out
    
    # helper function to display multiple examples in a row,
    # given an array of axes from pyplot,
    # and given a 1 by n matrix, where each row is an array of pixels,
    def plt_exp(axr, pxss):
        for ax, pxs in zip(axr, pxss):
            plt_dgt(ax, pxs)
    
    def create_examples(feature_log_prob, label, num_examples = 10):
        '''
        Helper function to generate a black or white pixel image feature vector.
        Given a feature probabilities list, and creates a black pixel, 
        if a random number is lower than the probability of a pixel being in that position,
        otherwise creates a white pixel
        '''
        new_examples = np.exp(feature_log_prob)
        new_examples = np.repeat([new_examples], num_examples, axis = 0)
        j = np.random.rand(num_examples, img_px*img_px)
        new_examples[new_examples < j] = 0
        new_examples[new_examples >= j] = 1
        return new_examples
    
    # creates and trains a bernouli naive bayes classifier
    clsfr = BernoulliNB(binarize = 0.0)
    clsfr.fit(mini_train_data, mini_train_labels)
    
    # for every label in the train set, i.e. 0 to 9, generates some examples
    num_examples = 20
    examples = dict()
    for label in range(0, 10):
        examples[label] = create_examples(clsfr.feature_log_prob_[label], label, num_examples)
    
    # plot our 10 by 20 matrix
    fig = plt.figure()
    fig, axs = plt.subplots(10, num_examples, figsize=(16,16))
    for r, pxss in zip(axs, examples.values()):
        plt_exp(r, pxss)

P10(20)
```

ANSWER: the generated digits are much more noisy vs the training digits

# 11

Remember that a strongly calibrated classifier is rougly 90% accurate when the posterior probability of the predicted class is 0.9. A weakly calibrated classifier is more accurate when the posterior is 90% than when it is 80%. A poorly calibrated classifier has no positive correlation between posterior and accuracy.

Train a BernoulliNB model with a reasonable alpha value. For each posterior bucket (think of a bin in a histogram), you want to estimate the classifier's accuracy. So for each prediction, find the bucket the maximum posterior belongs to and update the "correct" and "total" counters.

How would you characterize the calibration for the Naive Bayes model?

```python
def P11(buckets):
    
    #creates a bernoulli NB classifier, fits it to the mini train data, and makes predictions against the dev data
    clsfr = BernoulliNB(alpha=0.001, binarize=0.0)
    clsfr.fit(mini_train_data, mini_train_labels)
    preds = clsfr.predict(dev_data)
    preds_pp = clsfr.predict_proba(dev_data)
    
    # for every prediction, right or wrong, finds the posterior probability of that prediction
    pp = [preds_pp[idx, preds[idx].astype(int)] for idx, pd in enumerate(preds)]
    
    # for every bucket, tallies up the number correct, and the total
    pp_bins = np.digitize(pp, bins = buckets, right = True)
    total = np.bincount(pp_bins)
    crct = preds == dev_labels
    correct =  np.bincount(pp_bins, weights = crct)
    
    return correct, total

buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct, total = P11(buckets)

acrs = []
for i in range(len(buckets)):
    accuracy = 0.0
    if (total[i] > 0):
        accuracy = correct[i] / total[i]
        acrs.append(accuracy)
    else:
        acrs.append(accuracy)
    print('p(pred) <= %.13f    total = %3d    accuracy = %.3f' %(buckets[i], total[i], accuracy))

plt.plot(buckets, acrs)
plt.xlabel("p(pred)")
plt.ylabel("accuracy")
plt.title("Our BernoulliNB classifier is weakly calibrated")
```

ANSWER: Classifier BernoulliNB with binarize 0.0 and alpha 0.001 is weakly calibrated

```python
def P12():
    #the basic BNB with alpha 0.001 and binarize 0.0
    clsfr = BernoulliNB(alpha=0.001, binarize=0.0)
    clsfr.fit(mini_train_data, mini_train_labels)
    preds = clsfr.predict(dev_data)
    accuracy = clsfr.score(dev_data, dev_labels)
    print("accuracy of BNB:", accuracy)
    print(classification_report(preds, dev_labels))
    
    # helper function to display a single digit
    def plt_dgt(pxs, sh = (img_px, img_px)):
        pxs_r = pxs.reshape(sh)
        plt.figure(figsize=(3,3))
        plt.imshow(pxs_r, cmap='Greys')
        plt.show()
    
    # helper function to calculate features, column and rows sums
    def calc_col_row_sums(pxss):
        col_sums = []
        row_sums = []
        for i in range(0, len(pxss)):
            col_sums.append(pxss[i].reshape((img_px, img_px)).sum(axis = 0).tolist())
            row_sums.append(pxss[i].reshape((img_px, img_px)).sum(axis = 1).tolist())
        col_sums = np.array(col_sums)
        row_sums = np.array(row_sums)
        return col_sums, row_sums
    
    # helper function binarize the pixels
    # can't use binarize = 0.0 in classifier constructor because of extra features
    def bin_pxss(pxss, th = 0.0):
        pxs_b_w = np.copy(pxss)
        pxs_b_w[pxs_b_w <= th] = 0 # white
        pxs_b_w[pxs_b_w > th] = 1 # black
        return pxs_b_w
    
    # Helper function to calculate the number of enclosed regions
    # leverages the label function in the scikit learn image processing library to label areas of contiguous pixels
    # then subtracts 2 to calculate the number of enclosed regions.
    #
    # This function expectes binnarized input
    #
    # e.g. 0 has 1 region, 1 has 0 regions, 2 has 0 regions, 
    # 3 has 0 regions, 4 has 0 regions, 5 has 0 regions, 6 has 1 region, 
    # 7 has 0 regions, 8 has 2 regions, 9 has 1 regions 
    #
    def count_regions(pxss):
        from skimage.measure import label
        lb = label(pxss.reshape((img_px, img_px)), connectivity = 2, background = -1)
        num_regions = len(np.unique(lb)) - 2
        return num_regions
    
    # calculate more features on both train data, and dev data
    bin_train_data = bin_pxss(mini_train_data)
    bin_dev_data = bin_pxss(dev_data)
    
    t_col_sums, t_row_sums = calc_col_row_sums(mini_train_data)
    d_col_sums, d_row_sums = calc_col_row_sums(dev_data)
    
    t_num_regions = np.array([count_regions(pxss) for pxss in bin_train_data]).reshape(-1,1)
    d_num_regions = np.array([count_regions(pxss) for pxss in bin_dev_data]).reshape(-1,1)
    
    # concatenates the features
#     combo_train_data = np.concatenate((bin_train_data, t_col_sums, t_row_sums, t_num_regions), axis = 1)
#     combo_dev_data = np.concatenate((bin_dev_data, d_col_sums, d_row_sums, d_num_regions), axis = 1)
# 
#     combo_train_data = np.concatenate((bin_train_data, t_col_sums, t_num_regions), axis = 1)
#     combo_dev_data = np.concatenate((bin_dev_data, d_col_sums, d_num_regions), axis = 1)

    # ... for some reason, excluding the column sums makes model perform slightly better
    # maybe because the vertical axis of the digits was more stretched out, compared to the horizontal axis
    combo_train_data = np.concatenate((bin_train_data, t_row_sums, t_num_regions), axis = 1)
    combo_dev_data = np.concatenate((bin_dev_data, d_row_sums, d_num_regions), axis = 1)

    # BNB on more features, with alpha 0.001 and binarize 0.0 (sort of)
    clsfr = BernoulliNB(alpha=0.001)
    clsfr.fit(combo_train_data, mini_train_labels)
    preds = clsfr.predict(combo_dev_data)
    accuracy = clsfr.score(combo_dev_data, dev_labels)
    print("accuracy of BNB on more features:", accuracy)
    print(classification_report(preds, dev_labels))
    

P12()
```
